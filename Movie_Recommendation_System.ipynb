```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python

data = {
    'movie_id': [1, 2, 3, 4, 5],
    'title': ['The Dark Knight', 'Inception', 'Interstellar', 'The Hangover', 'Superbad'],
    'genres': ['Action Thriller Dark', 'Sci-Fi Thriller Mind-bending', 'Sci-Fi Space Drama', 'Comedy Title Funny', 'Comedy Teen Funny']
}

df = pd.DataFrame(data)
print("Dataset:")
print(df)
```

```python
# TF-IDF Vectorizer initialize
tfidf = TfidfVectorizer(stop_words='english')


tfidf_matrix = tfidf.fit_transform(df['genres'])

print("Matrix Shape:", tfidf_matrix.shape)
```

```python

cosine_sim = cosine_similarity(tfidf_matrix, tfidf_matrix)

print("Similarity Matrix (Sample):")
print(cosine_sim)
```

```python
def get_recommendations(movie_title, cosine_sim=cosine_sim):

    try:
        idx = df[df['title'] == movie_title].index[0]
    except IndexError:
        return f"Sorry, '{movie_title}' humare database mein nahi hai."


    sim_scores = list(enumerate(cosine_sim[idx]))

    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)


    sim_scores = sim_scores[1:3]


    movie_indices = [i[0] for i in sim_scores]


    return df['title'].iloc[movie_indices].tolist()
```

```python
# Output:
print("Recommendations for 'Inception':")
print(get_recommendations('Inception'))

print("\nRecommendations for 'The Hangover':")
print(get_recommendations('The Hangover'))
```
